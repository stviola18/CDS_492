```python
#This simulation prints out the year and how many meteorites fell, followed by the mass of each
import random

def simulate_meteorite(years):
    # Percentage of total meteorites by mass group. Data obtained in R
    meteor_percentage = {"Pebble": 0.845185066,
                         "Very Small": 0.045969805,
                         "Small": 0.077391444,
                         "Medium": 0.021022272,
                         "Large": 0.009004967,
                         "Very Large": 0.001506169,
                         "Enormous": 0.000192277}
    
    # Meteorite sizes in each mass group
    meteor_sizes = {"Pebble": [0, 1],
                    "Very Small": [1, 2],
                    "Small": [2, 20],
                    "Medium": [20, 100],
                    "Large": [100, 1000],
                    "Very Large": [1000, 10000],
                    "Enormous": [10000, 10000000]}
    
    # Rate per year that meteorites land on Earth. Data obtained in R
    meteor_rate = 15.50173869846
    
    # list of count of meteorites that fall per year
    meteor_count = {year: 0 for year in range(years)}
    
    # list of the size of each meteorite that falls each year
    meteor_size = {year: [] for year in range(years)}
    
    # Simulates the meteorite falls for the given number of years
    for year in range(years):
        num_meteors = int(meteor_rate + random.gauss(0, 1))
        meteor_count[year] = num_meteors
        
        # Simulates the size of each meteorite that falls in this year
        for i in range(num_meteors):
            mass_group = random.choices(list(meteor_percentage.keys()),
                                        weights=list(meteor_percentage.values()),
                                        k=1)[0]
            
            # Randomly selects a size of the meteorite
            meteor_mass = random.uniform(*meteor_sizes[mass_group])
            meteor_size[year].append(meteor_mass)
    
    # Prints a breakdown of  the amount and the mass of every meteorite that hit per year
    for year in range(years):
        print(f"Year {year + 2024}: {meteor_count[year]} meteorites fell to Earth")
        print(f"Size of each meteor(lbs): {meteor_size[year]}")



simulate_meteorite(10)
```

```python
#This simulation creates a dataset which will be used for validation
import random
import pandas as pd

def simulate_meteorite_dataset74(years):
    # Percentage of total meteorites by mass group
    meteor_percentage = {"Pebble": 0.880897583429229,
                         "Very Small": 0.043457659243214,
                         "Small": 0.0616665538482366,
                         "Medium": 0.0101875042306911,
                         "Large": 0.00338455290056184,
                         "Very Large": 0.00040614634806742,
                         "Enormous": 0.000000000001}
    
    # Meteorite sizes in each mass group
    meteor_sizes = {"Pebble": [0, 1],
                    "Very Small": [1, 2],
                    "Small": [2, 20],
                    "Medium": [20, 100],
                    "Large": [100, 1000],
                    "Very Large": [1000, 10000],
                    "Enormous": [10000, 10000000]}
    
    # Rate per year that meteorites land on Earth
    meteor_rate = 757.589743589744
    
    # Create a list to hold the meteorite dataframes for each year
    meteor_data_list = []
    
    # Simulates the meteorite falls for the given number of years
    for year in range(years):
        num_meteors = int(meteor_rate + random.gauss(0, 1))
        
        # Simulates the size of each meteorite that falls in this year
        year_data = {"year": [year + 1974] * num_meteors,
                     "mass": [random.uniform(*meteor_sizes[random.choices(list(meteor_percentage.keys()),
                                                                            weights=list(meteor_percentage.values()),
                                                                            k=1)[0]]) for i in range(num_meteors)]}
        
        # Append the year's meteorite data to the list
        meteor_data_list.append(pd.DataFrame(year_data))
    
    # Concat the list of dataframes into a single dataframe
    meteor_data = pd.concat(meteor_data_list)
    
    meteor_data.to_csv('meteor_sim_data.csv', index = False)
    
    return meteor_data
simulate_meteorite_dataset74(39)
```
